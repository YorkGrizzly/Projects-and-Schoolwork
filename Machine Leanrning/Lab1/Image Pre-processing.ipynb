```python
def display_one(a, title1="Original"):
    plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def display(a, b, c, title1="Original", title2="Edited1", title3 = "Edited2"):
    plt.subplot(131), plt.imshow(a), plt.title(title1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(132), plt.imshow(b), plt.title(title2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(133), plt.imshow(c), plt.title(title3)
    plt.xticks([]), plt.yticks([])
    plt.show()
    
def preprocessImg(img, new_width=30, new_height=30):
    new_dim = (new_width, new_height)
    return cv2.resize(img, new_dim, interpolation=cv2.INTER_LINEAR)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mping
import os
import cv2

imgSet = []
raw_images = []

path = './sum/green/'
files = os.listdir(path)
#train_green = []

for file in files: 
    p = path + file
    raw_img = cv2.imread(p)
    raw_images.append(raw_img)
    img = preprocessImg(raw_img)
    imgSet.append(img)

    #display_one(img)
    
# Gaussian
no_noise = []
for i in range(len(imgSet)):
    blur = cv2.GaussianBlur(raw_images[i], (5, 5), 0)
    no_noise.append(blur)
    
display_one(raw_images[19])

path = './sum/red/'
files = os.listdir(path)


for file in files:
    p = path + file
    img = cv2.imread(p)
    raw_images.append(img)

    resized_img = preprocessImg(img)
    blurred_img = cv2.GaussianBlur(resized_img, (5, 5), 0) # apply Gausian blur
    imgSet.append(resized_img)

display_one(raw_images[42])
```
